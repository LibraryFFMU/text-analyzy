# Textová analýza sylabu předmětů

Pro větší zacílení lekcí informačního vzdělávání, i v oblasti specializovaných oborových lekcí sledujeme, které předměty na fakultě jsou věnovány příbuzným tématům. Pro tento účel definuje příslušné oddělení klíčová slova, která je třeba vyhledat v sylabu předmětů.

vstupní data:
- seznam klíčových slov
- export sylabu předmětů dané fakulty za požadované období z informačního systému univerzity

výstup:
- xlsx soubor obsahující seznam předmětů ve struktuře - kód předmětu, název předmětu, dodavatelské pracoviště, vyučující   )

```python
import pandas as pd
import numpy as np 

#export předmětového katalogu z informačnícho systému školy
df = pd.read_excel("FF-export_katalog-20_2021.xlsx")
df.head(1)
```

```python
#výběr potřebných sloupců pro další práci, náš export na vstupu obsahuje zbytečně moc sloupců
vyber = df[["kód", "název", "dodavatelské pracoviště","přednášející", "cíle","metody hodnocení", "výukové metody", "osnova"]]
vyber.head(2)
```

```python
#vytvoření seznamu hledaných slov
slova = ['informa', 'elektronick', 'zdroj', 'publika', 'pramen',
         'odborn', 'člán', 'text', 'cita', 'citovan', 'psaní',
         'diplom', 'literatur', 'vědeck', 'referát', 'prezent' ]
```

```python
# vytvoření souboru pro zápis výstupu, prohlednání sloupců ve smyčce
# smyčka prochází všechny řádky - vybrané sloupce spojuje do textu a v něm hledá klíčová slova

fo = open("vystup.txt", mode="w", encoding='utf8')

for i, line in vyber.iterrows():
  txt = [ str(line['název']), str(line['cíle']), str(line['metody hodnocení']),
          str(line['výukové metody']), str(line['osnova'])
        ]
  txt = ' '.join(txt)
  # print(txt)
  found = []
  for word in slova:
    if txt.find(word) >= 0:
      found.append(word) 
  # print(found)
  fo.write(str(line['kód']) + '\t' + str(line['název']) + '\t' + 
           str(line['dodavatelské pracoviště']) + '\t' + 
           str(line['přednášející']) + '\t' + ','.join(found) + '\n')
  # break

fo.close()
print('ok')
```
